```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Define the data as a dictionary of pandas DataFrames
data = {
    'Overall': pd.DataFrame({
        'Player': ['Watkins', 'Saka', 'Salah', 'Son', 'Foden'],
        'Points': [185, 178, 168, 166, 160],
        'Price': [9.0, 9.1, 13.0, 10.0, 8.2]
    }).sort_values('Points', ascending=False),
    'Forwards': pd.DataFrame({
        'Player': ['Watkins', 'Haaland', 'Solanke', 'J. Alvarez', 'Morris'],
        'Points': [185, 154, 138, 130, 113],
        'Price': [9.0, 14.3, 7.2, 6.5, 5.2]
    }).sort_values('Points', ascending=False),
    # ... (other categories)
}

# Set the aesthetic style of the plots
sns.set(style='white', context='talk')

# Function to create horizontal bar charts for points
def create_horizontal_bar_chart(category, df):
    # Initialize the matplotlib figure
    f, ax = plt.subplots(figsize=(6, 4))  # Smaller figure size

    # Plot the Points
    bars = sns.barplot(x='Points', y='Player', data=df,
                label='Total', color='b', edgecolor='w')

    # Display the point values on the bars
    for bar in bars.patches:
        ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
                f'{bar.get_width()}',
                ha='center', va='center', fontsize=10, color='black')

    # Remove the legend, x-scale, and spines
    ax.get_legend().remove()
    ax.get_xaxis().set_visible(False)
    sns.despine(left=True, bottom=True)

    # Set the title
    plt.title(f'{category} - Fantasy Premier League Gameweek 29')

    plt.show()

# Generate horizontal bar charts for each category
for category, df in data.items():
    create_horizontal_bar_chart(category, df)
```
