```python
import requests, json
from pprint import pprint
base_url = 'https://fantasy.premierleague.com/api/' 
r = requests.get(base_url+'bootstrap-static/').json()
r
```

```python
import pandas as pd

df = pd.json_normalize(r)
df
```

```python
import json

r_strings  = json.dumps(r)
teams_df = pd.json_normalize(json.loads(r_strings), 'teams', sep='_')

teams_df = teams_df.sort_values(by='name')
teams_df
```

```python
positions_df = pd.json_normalize(json.loads(r_strings), 'element_types', sep='_')
positions_df
```

```python
players_df = pd.json_normalize(json.loads(r_strings), 'elements', sep='_')
players_df = players_df.sort_values(by='team')
players_df
```

```python
cols = list(players_df)
cols
```

```python
cols.insert(0, cols.pop(cols.index('first_name')))
cols.insert(1, cols.pop(cols.index('second_name')))
cols.insert(2, cols.pop(cols.index('web_name')))
cols.insert(3, cols.pop(cols.index('now_cost')))

players_df = players_df[cols]
players_df
```

```python
players_df.to_csv('player_data.csv', index=False)
positions_df.to_csv('position_lookup.csv', index=False)
teams_df.to_csv('teams_lookup.csv', index=False)
```

```python
selected_columns = ['web_name', 'now_cost', 'total_points','ict_index','team', 'points_per_game','starts', 'element_type', 'value_season']
selected_data = players_df[selected_columns]
player_json = selected_data.to_dict(orient='records')
player_json
```
